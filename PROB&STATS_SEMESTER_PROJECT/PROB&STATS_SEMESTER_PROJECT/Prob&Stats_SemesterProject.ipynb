```python
# =============================================================================
#  INTERCITY BUS TRAVEL DELAY PREDICTOR
#  Probability & Statistics Course Project
# =============================================================================
#  GROUP MEMBERS:
#
#  24F-0008: Eesha Saqib
#  24F-0001: Rameen Ahmad
#  24F-0633: Subhan Ahmad
#  24F-3018: Fatima Mobasher
#  24P-0696: Haider Sultan
# =============================================================================

print("BBBBBBBBBBB       UU         UU      SSSSSSSSSS            DDDDDDDDDDDD       EEEEEEEEEEEE   LL                    A        YY          YY")
print("BB        BB      UU         UU     SS        SS           DD          DD     EE        EE   LL                   AAA        YY        YY")
print("BB         BB     UU         UU    SS          SS          DD           DD    EE             LL                  AA AA        YY      YY")
print("BB         BB     UU         UU    SS          SS          DD            DD   EE             LL                 AA   AA        YY    YY")
print("BB        BB      UU         UU     SS                     DD            DD   EE             LL                AA     AA        YY  YY")
print("BBBBBBBBBBB       UU         UU      SSSSSSSSSS            DD            DD   EEEEEEEEEEEE   LL               AAAAAAAAAA         YYYY")
print("BB         BB     UU         UU               SS           DD            DD   EE             LL              AA         AA        YY ")
print("BB          BB    UU         UU    SS          SS          DD            DD   EE             LL             AA           AA       YY")
print("BB          BB    UU         UU    SS          SS          DD           DD    EE             LL             AA           AA       YY")
print("BB         BB      UU       UU      SS        SS           DD          DD     EE         EE  LLLLLLLLLLLL   AA           AA       YY")
print("BBBBBBBBBBBB        UUUUUUUUU        SSSSSSSSSS            DDDDDDDDDDDD       EEEEEEEEEEEEE  LLLLLLLLLLLL   Aa           AA       YY")

import warnings, sys
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from scipy import stats
from scipy.stats import (norm, lognorm, expon, gamma,
                         ttest_ind, f_oneway, chi2_contingency)
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.regression.linear_model import OLS
from statsmodels.tools import add_constant
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import RidgeCV, LassoCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from xgboost import XGBRegressor
from fitter import Fitter

plt.rcParams.update({"figure.dpi": 110, "axes.spines.top": False,
                     "axes.spines.right": False, "font.size": 11})
sns.set_palette("husl")

# =============================================================================
# LOAD DATASET
# =============================================================================
XLSX_PATH = "Pakistan_Bus_Delay_Dataset.xlsx"

df = pd.read_excel(XLSX_PATH, sheet_name="Bus_Delay_Data", engine="openpyxl")

# mapping logical names to actual column headers in the spreadsheet
COL = {
    "delay"    : "Delay_Minutes",
    "weather"  : "Weather",
    "day"      : "Day",
    "distance" : "Distance_KM",
    "departure": "Departure_Hour",
    "event"    : "Special_Event",
    "route"    : "Route",
    "bus_type" : "Bus_Type",
    "road_type": "Road_Type",
    "traffic"  : "Traffic_Level",
}

for k, v in COL.items():
    assert v in df.columns, f"Column '{v}' not found for key '{k}'"
TARGET = COL["delay"]

df[TARGET] = pd.to_numeric(df[TARGET], errors="coerce")
df.dropna(subset=[TARGET], inplace=True)
df = df[df[TARGET] >= 0].reset_index(drop=True)


# =============================================================================
# EDA 1 — STRUCTURAL OVERVIEW
# =============================================================================
print("\n" + "="*65)
print("EDA 1 — STRUCTURAL OVERVIEW")
print("="*65)

print(f"\n  Shape              : {df.shape[0]:,} rows  x  {df.shape[1]} columns")
print(f"  Memory usage       : {df.memory_usage(deep=True).sum()/1024:.1f} KB")

dtype_series = df.dtypes.astype(str).value_counts()
print(f"\n  Column dtype breakdown:\n{dtype_series.to_string()}")

col_info = pd.DataFrame({
    "dtype"    : df.dtypes,
    "non_null" : df.notnull().sum(),
    "null_pct" : (df.isnull().mean()*100).round(2),
    "nunique"  : df.nunique(),
    "sample"   : df.iloc[0],
})
print(f"\n  Per-column audit:\n{col_info.to_string()}")

n_dupes = int(df.duplicated().sum())
print(f"\n  Duplicate rows : {n_dupes}")
if n_dupes > 0:
    df.drop_duplicates(inplace=True)
    df.reset_index(drop=True, inplace=True)
    print(f"  Dropped duplicates. New shape: {df.shape}")

print(f"\n  Full numeric summary:\n{df.describe().T.to_string()}")


# =============================================================================
# EDA 2 — MISSING VALUE ANALYSIS
# =============================================================================
print("\n" + "="*65)
print("EDA 2 — MISSING VALUE ANALYSIS")
print("="*65)

null_counts = df.isnull().sum()
null_pct    = (df.isnull().mean() * 100).round(2)
null_report = pd.DataFrame({"Missing #": null_counts, "Missing %": null_pct})

if null_report["Missing #"].sum() == 0:
    print("\nZero missing values — dataset is complete.")
else:
    null_report = (null_report[null_report["Missing #"] > 0]
                   .sort_values("Missing %", ascending=False))
    print(f"\n{null_report.to_string()}")
    fig, ax = plt.subplots(figsize=(12, 4))
    sns.heatmap(df.sample(200, random_state=1).isnull(),
                cbar=False, yticklabels=False, ax=ax, cmap="YlOrRd")
    ax.set_title("Missing Value Heatmap (200-row sample)")
    plt.tight_layout(); plt.savefig("eda_02_missing.png"); plt.show()


# =============================================================================
# EDA 3 — TARGET VARIABLE DEEP DIVE
# =============================================================================
print("\n" + "="*65)
print("EDA 3 — TARGET VARIABLE: Delay Minutes")
print("="*65)

y = df[TARGET]

print(f"\n  pandas .describe():\n{y.describe().to_string()}")

Q1, Q3 = y.quantile(0.25), y.quantile(0.75)
IQR    = Q3 - Q1

# collecting all the key descriptive measures in one place
extra = pd.Series({
    "Mean"      : y.mean(),
    "Median"    : y.median(),
    "Mode"      : y.mode().iloc[0],
    "Std Dev"   : y.std(),
    "Variance"  : y.var(),
    "Skewness"  : y.skew(),
    "Kurtosis"  : y.kurtosis(),
    "IQR"       : IQR,
    "CV (%)"    : y.std() / y.mean() * 100,
    "Range"     : y.max() - y.min(),
    "Min"       : y.min(),
    "Max"       : y.max(),
})
print(f"\n  Extended descriptive stats:")
for k, v in extra.items():
    print(f"    {k:<14}  {v:>10.4f}")

pct_levels  = np.array([1, 5, 10, 25, 50, 75, 90, 95, 99])
percentiles = pd.Series(
    np.percentile(y.values, pct_levels),
    index=[f"P{p:02d}" for p in pct_levels],
    name="Delay (min)"
)
print(f"\n  Percentile table:")
print(percentiles.to_string())

z_scores = np.abs(stats.zscore(y.values))
z_mask   = z_scores > 3
iqr_mask = (y < Q1 - 1.5*IQR) | (y > Q3 + 1.5*IQR)

print(f"\n  Outliers | Z>3     : {z_mask.sum():>5}  ({z_mask.mean()*100:.2f}%)")
print(f"  Outliers | IQR     : {iqr_mask.sum():>5}  ({iqr_mask.mean()*100:.2f}%)")
if z_mask.sum() > 0:
    print(f"  Z-outlier range    : {y[z_mask].min():.1f} - {y[z_mask].max():.1f} min")

fig = plt.figure(figsize=(15, 10))
fig.suptitle("EDA 3 — Delay Distribution Analysis", fontsize=13,
             fontweight="bold", y=1.01)
gs  = gridspec.GridSpec(2, 2, figure=fig, hspace=0.38, wspace=0.32)

ax0 = fig.add_subplot(gs[0, 0])
ax0.hist(y, bins=60, density=True, color="#5B8DB8", edgecolor="white", alpha=0.85)
yk  = np.linspace(y.min(), y.max(), 300)
ax0.plot(yk, stats.gaussian_kde(y)(yk), color="#E07B54", lw=2, label="KDE")
ax0.axvline(y.mean(),   color="crimson", lw=1.8, ls="--",
            label=f"Mean = {y.mean():.1f}")
ax0.axvline(y.median(), color="green",   lw=1.8, ls=":",
            label=f"Median = {y.median():.1f}")
ax0.set(title="Histogram + KDE", xlabel="Delay (min)", ylabel="Density")
ax0.legend(fontsize=8)

ax1  = fig.add_subplot(gs[0, 1])
logy = np.log1p(y)
lx   = np.linspace(logy.min(), logy.max(), 300)
ax1.hist(logy, bins=50, color="#A78BFA", edgecolor="white", density=True, alpha=0.85)
ax1.plot(lx, stats.gaussian_kde(logy)(lx), color="#E07B54", lw=2)
ax1.set(title="Histogram — log(1+Delay)",
        xlabel="log(1+Delay)", ylabel="Density")

ax2 = fig.add_subplot(gs[1, 0])
ax2.boxplot(y, vert=True, patch_artist=True, notch=True,
            boxprops=dict(facecolor="#5B8DB8", alpha=0.6),
            medianprops=dict(color="red", lw=2),
            flierprops=dict(marker=".", markersize=2, alpha=0.3))
ax2.set(title=f"Boxplot (Tukey fences) — {iqr_mask.sum()} fliers",
        ylabel="Delay (min)")
ax2.text(1.18, float(y.median()), f"  Median\n  {y.median():.1f} min",
         va="center", fontsize=8, color="red")

ax3      = fig.add_subplot(gs[1, 1])
sorted_y = np.sort(y.values)
ecdf     = np.arange(1, len(sorted_y)+1) / len(sorted_y)
ax3.plot(sorted_y, ecdf, color="#5B8DB8", lw=1.5, label="ECDF")
for pv, lbl in [(0.50,"P50"), (0.90,"P90"), (0.95,"P95")]:
    xp = np.percentile(sorted_y, pv*100)
    ax3.axvline(xp, ls="--", lw=1, alpha=0.7)
    ax3.text(xp+0.5, pv-0.05, f"{lbl}={xp:.0f}m", fontsize=8)
ax3.set(title="ECDF", xlabel="Delay (min)", ylabel="Cumulative Probability")
plt.savefig("eda_03_target.png", bbox_inches="tight"); plt.show()


# =============================================================================
# EDA 4 — CATEGORICAL FEATURE ANALYSIS
# =============================================================================
print("\n" + "="*65)
print("EDA 4 — CATEGORICAL FEATURE ANALYSIS")
print("="*65)

cat_cols = [v for k, v in COL.items()
            if v and k in ("weather","day","bus_type","road_type","traffic","route")
            and df[v].dtype == object]

if cat_cols:
    for col in cat_cols:
        grp = (df.groupby(col)[TARGET]
                 .agg(Count="count", Mean="mean", Median="median",
                      Std="std", Min="min", Max="max")
                 .round(2)
                 .sort_values("Median", ascending=False))
        print(f"\n  Delay stats grouped by '{col}':\n{grp.to_string()}")

    print("\n  Category frequency (% share):")
    for col in cat_cols:
        vc = (df[col].value_counts(normalize=True).mul(100).round(1))
        print(f"\n    {col}:\n{vc.to_string()}")

    n = len(cat_cols)
    fig, axes = plt.subplots(1, n, figsize=(6*n, 5), squeeze=False)
    fig.suptitle("EDA 4A — Delay Boxplot by Category", fontsize=13, fontweight="bold")
    palette = sns.color_palette("husl", 8)
    for i, col in enumerate(cat_cols):
        ax    = axes[0][i]
        order = (df.groupby(col)[TARGET].median()
                   .sort_values(ascending=False).index.tolist())
        sns.boxplot(data=df, x=col, y=TARGET, order=order,
                    palette=palette, ax=ax,
                    flierprops=dict(marker=".", markersize=2, alpha=0.3))
        medians = df.groupby(col)[TARGET].median().reindex(order)
        for j, med in enumerate(medians):
            ax.text(j, med+0.5, f"{med:.1f}", ha="center",
                    fontsize=8, fontweight="bold")
        ax.set(title=f"Delay by {col}", xlabel=col, ylabel="Delay (min)")
        ax.tick_params(axis="x", rotation=30)
    plt.tight_layout(); plt.savefig("eda_04a_boxplot.png"); plt.show()

    if COL["weather"] and COL["day"]:
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))
        fig.suptitle("EDA 4B — Violin Plots", fontsize=13, fontweight="bold")
        for ax, col in zip(axes, [COL["weather"], COL["day"]]):
            order = (df.groupby(col)[TARGET].median()
                       .sort_values(ascending=False).index.tolist())
            sns.violinplot(data=df, x=col, y=TARGET, order=order,
                           palette="husl", inner="quartile", ax=ax)
            ax.set(title=f"Violin — Delay by {col}",
                   xlabel=col, ylabel="Delay (min)")
            ax.tick_params(axis="x", rotation=30)
        plt.tight_layout(); plt.savefig("eda_04b_violin.png"); plt.show()


# =============================================================================
# EDA 5 — NUMERICAL FEATURE ANALYSIS
# =============================================================================
print("\n" + "="*65)
print("EDA 5 — NUMERICAL FEATURE ANALYSIS")
print("="*65)

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
num_cols = [c for c in num_cols if c != TARGET]

if num_cols:
    num_summary = df[num_cols + [TARGET]].describe().T
    num_summary["skew"]     = df[num_cols + [TARGET]].skew()
    num_summary["kurtosis"] = df[num_cols + [TARGET]].kurtosis()
    print(f"\n  Numeric feature summary:\n{num_summary.round(3).to_string()}")

    corr_matrix = df[num_cols + [TARGET]].corr(method="pearson")
    target_corr = (corr_matrix[TARGET].drop(TARGET)
                                      .sort_values(key=abs, ascending=False))
    print(f"\n  Pearson correlation with target '{TARGET}':")
    print(target_corr.round(4).to_string())

    spear_corr = df[num_cols + [TARGET]].corr(method="spearman")[TARGET].drop(TARGET)
    print(f"\n  Spearman correlation with target:")
    print(spear_corr.round(4).to_string())

    fig, ax = plt.subplots(figsize=(max(6, len(num_cols)),
                                    max(5, len(num_cols)-1)))
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f",
                cmap="coolwarm", center=0, linewidths=0.5,
                annot_kws={"size": 9}, ax=ax)
    ax.set_title("EDA 5A — Pearson Correlation Heatmap", fontsize=12, fontweight="bold")
    plt.tight_layout(); plt.savefig("eda_05a_corr.png"); plt.show()

    n = len(num_cols)
    if n > 0:
        fig, axes = plt.subplots(1, n, figsize=(6*n, 4), squeeze=False)
        fig.suptitle("EDA 5B — Numeric Features vs Delay", fontsize=13, fontweight="bold")
        for i, col in enumerate(num_cols):
            ax    = axes[0][i]
            clean = df[[col, TARGET]].dropna()
            ax.scatter(clean[col], clean[TARGET], alpha=0.25, s=8, color="#5B8DB8")
            m, b  = np.polyfit(clean[col].values, clean[TARGET].values, 1)
            xline = np.linspace(clean[col].min(), clean[col].max(), 200)
            r     = clean[col].corr(clean[TARGET])
            ax.plot(xline, m*xline+b, color="crimson", lw=1.5, label=f"r = {r:.3f}")
            ax.set(title=f"{col} vs Delay", xlabel=col, ylabel="Delay (min)")
            ax.legend(fontsize=9)
        plt.tight_layout(); plt.savefig("eda_05b_scatter.png"); plt.show()


# =============================================================================
# EDA 6 — PIVOT TABLE ANALYSIS
# =============================================================================
print("\n" + "="*65)
print("EDA 6 — PIVOT TABLE ANALYSIS")
print("="*65)

if COL["weather"] and COL["day"]:
    pivot_mean = pd.pivot_table(df, values=TARGET,
                                index=COL["weather"],
                                columns=COL["day"],
                                aggfunc="mean").round(1)
    pivot_n = pd.pivot_table(df, values=TARGET,
                              index=COL["weather"],
                              columns=COL["day"],
                              aggfunc="count")
    print(f"\n  Mean delay — Weather x Day:\n{pivot_mean.to_string()}")
    print(f"\n  Sample count — Weather x Day:\n{pivot_n.to_string()}")

    fig, axes = plt.subplots(1, 2, figsize=(14, 4))
    sns.heatmap(pivot_mean, annot=True, fmt=".1f", cmap="YlOrRd",
                linewidths=0.4, ax=axes[0])
    axes[0].set_title("Mean Delay — Weather x Day")
    sns.heatmap(pivot_n, annot=True, fmt=".0f", cmap="Blues",
                linewidths=0.4, ax=axes[1])
    axes[1].set_title("Sample Count — Weather x Day")
    plt.tight_layout(); plt.savefig("eda_06_pivot.png"); plt.show()

if COL["departure"] and COL["day"]:
    # bin numeric departure hour into four broad time periods
    df["_dep_period"] = pd.cut(
        df[COL["departure"]],
        bins=[-1, 5, 11, 17, 23],
        labels=["Night(0-5)", "Morning(6-11)", "Afternoon(12-17)", "Evening(18-23)"]
    )
    pivot2 = pd.pivot_table(df, values=TARGET,
                             index="_dep_period",
                             columns=COL["day"],
                             aggfunc=np.median).round(1)
    print(f"\n  Median delay — Departure Period x Day:\n{pivot2.to_string()}")


# =============================================================================
# EDA 7 — DELAY BINNING AND FREQUENCY TABLE
# =============================================================================
print("\n" + "="*65)
print("EDA 7 — DELAY BINNING AND FREQUENCY TABLE")
print("="*65)

bins   = [-np.inf, 0, 5, 15, 30, 60, np.inf]
labels = ["On time","Slight (1-5)","Minor (6-15)",
          "Moderate (16-30)","Severe (31-60)","Extreme (60+)"]
df["delay_category"] = pd.cut(df[TARGET], bins=bins, labels=labels)

freq     = df["delay_category"].value_counts().reindex(labels).rename("Count")
freq_pct = (freq / freq.sum() * 100).round(2).rename("Pct (%)")
freq_cum = freq_pct.cumsum().round(2).rename("Cum. (%)")
freq_tbl = pd.concat([freq, freq_pct, freq_cum], axis=1)
print(f"\n  Delay frequency table:\n{freq_tbl.to_string()}")

if COL["weather"]:
    ct = pd.crosstab(df["delay_category"], df[COL["weather"]],
                     normalize="index").mul(100).round(1)
    print(f"\n  Delay category x Weather (row-normalised %):\n{ct.to_string()}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("EDA 7 — Delay Categories", fontsize=13, fontweight="bold")
colors = ["#2ecc71","#f1c40f","#e67e22","#e74c3c","#8e44ad","#2c3e50"]
freq.plot.bar(ax=axes[0], color=colors, edgecolor="white")
for p in axes[0].patches:
    axes[0].text(p.get_x()+p.get_width()/2, p.get_height()+5,
                 f"{p.get_height():,.0f}", ha="center", fontsize=8)
axes[0].set(title="Delay Category Frequency", xlabel="", ylabel="Count")
axes[0].tick_params(axis="x", rotation=30)
if COL["weather"] and not ct.empty:
    sns.heatmap(ct, annot=True, fmt=".1f", cmap="YlOrRd",
                linewidths=0.4, ax=axes[1])
    axes[1].set_title("Delay Category x Weather (row %)")
    axes[1].tick_params(axis="x", rotation=30)
else:
    axes[1].axis("off")
plt.tight_layout(); plt.savefig("eda_07_bins.png"); plt.show()


# =============================================================================
# EDA 8 — ROUTE-LEVEL ANALYSIS
# =============================================================================
if COL["route"]:
    print("\n" + "="*65)
    print("EDA 8 — ROUTE-LEVEL ANALYSIS")
    print("="*65)

    route_stats = (df.groupby(COL["route"])[TARGET]
                     .agg(trips="count",
                          mean_delay="mean",
                          median_delay="median",
                          std_delay="std",
                          p90_delay=lambda x: np.percentile(x.values, 90))
                     .round(2)
                     .sort_values("mean_delay", ascending=False))

    print(f"\n  Top 10 worst routes (mean delay):")
    print(route_stats.nlargest(10, "mean_delay").to_string())
    print(f"\n  Top 10 most reliable routes (lowest median):")
    print(route_stats.nsmallest(10, "median_delay").to_string())

    top15 = route_stats.nlargest(15, "mean_delay")
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    fig.suptitle("EDA 8 — Route-Level Delay Profile", fontsize=13, fontweight="bold")
    top15["mean_delay"].sort_values().plot.barh(ax=axes[0], color="#E07B54", edgecolor="white")
    axes[0].set(title="Top 15 Worst Routes (mean delay)", xlabel="Mean Delay (min)")
    top15["std_delay"].reindex(top15.index).sort_values().plot.barh(ax=axes[1], color="#A78BFA", edgecolor="white")
    axes[1].set(title="Delay Variability (std) — same routes", xlabel="Std Dev (min)")
    plt.tight_layout(); plt.savefig("eda_08_routes.png"); plt.show()


# =============================================================================
# EDA 9 — PAIRPLOT
# =============================================================================
print("\n" + "="*65)
print("EDA 9 — PAIRPLOT")
print("="*65)

pair_cols = [v for v in [COL["distance"], TARGET] if v]
if COL["weather"] and len(pair_cols) >= 2:
    pair_df = (df[pair_cols + [COL["weather"]]]
                 .dropna()
                 .sample(min(1500, len(df)), random_state=42))
    g = sns.pairplot(pair_df, hue=COL["weather"], diag_kind="kde",
                     plot_kws={"alpha": 0.35, "s": 15}, height=3)
    g.fig.suptitle("EDA 9 — Pairplot (n=1500 sample)", y=1.02,
                   fontsize=12, fontweight="bold")
    plt.savefig("eda_09_pairplot.png", bbox_inches="tight"); plt.show()
    print("  Pairplot saved.")
else:
    print("  Skipped — weather or distance column not found.")

print("\nEDA COMPLETE — all plots saved.\n")


# =============================================================================
# PROB 1 — DISTRIBUTION FITTING
# =============================================================================
print("="*65)
print("PROB 1 — DISTRIBUTION FITTING  (Kolmogorov-Smirnov + fitter)")
print("="*65)

y_pos_all = y[y > 0]
sample    = y_pos_all.sample(min(3000, len(y_pos_all)), random_state=42).values
f         = Fitter(sample, distributions=["norm","lognorm","gamma","beta","uniform"])
f.fit()
best_dist_name = list(f.get_best().keys())[0]
print(f"\n  Best-fit distribution : {best_dist_name.upper()}")
print(f.summary())

for dist_name in ["norm", "lognorm", "gamma"]:
    dist      = getattr(stats, dist_name)
    params    = dist.fit(sample)
    ks, p_val = stats.kstest(sample, dist_name, args=params)
    print(f"  KS({dist_name:<9}): stat={ks:.4f}  p={p_val:.4f}  "
          f"{'good fit' if p_val>0.05 else 'rejected'}")

fig, ax = plt.subplots(figsize=(5, 5))
stats.probplot(y, dist="norm", plot=ax)
ax.set_title("QQ Plot — Delay_Minutes vs Normal")
plt.tight_layout(); plt.savefig("prob_01_qqplot.png"); plt.show()


# =============================================================================
# PROB 2 — BAYES, CONDITIONAL PROBABILITY, CLT, MONTE CARLO
# =============================================================================
print("\n" + "="*65)
print("PROB 2 — BAYES + CONDITIONAL PROBABILITY + CLT + MONTE CARLO")
print("="*65)

THRESHOLD     = 15
df["is_late"] = (df[TARGET] > THRESHOLD).astype(int)
p_late        = df["is_late"].mean()
print(f"\n  P(late > {THRESHOLD} min) = {p_late:.4f}  [{p_late*100:.1f}%]")

if COL["weather"]:
    cond = (df.groupby(COL["weather"])["is_late"]
              .agg(["mean","count"])
              .rename(columns={"mean": "P(late|weather)", "count": "n"}))
    print(f"\n  Conditional probability table:\n{cond.to_string()}")

    # verify Bayes theorem gives the same result as direct calculation
    for w in df[COL["weather"]].dropna().unique():
        p_w      = (df[COL["weather"]] == w).mean()
        p_w_late = df.loc[df["is_late"]==1, COL["weather"]].eq(w).mean()
        bayes    = p_w_late * p_late / p_w if p_w > 0 else 0
        direct   = df.loc[df[COL["weather"]]==w, "is_late"].mean()
        match    = "OK" if abs(bayes-direct) < 0.001 else "check"
        print(f"    {str(w):<15} Bayes={bayes:.4f}  Direct={direct:.4f}  {match}")

# CLT: even with a non-normal delay distribution, sample means converge to normal
sample_means = [y.sample(40, replace=True).mean() for _ in range(2000)]
fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(sample_means, bins=40, density=True, color="#A78BFA",
        edgecolor="white", label="Sample means (n=40, 2000 iters)")
xr = np.linspace(min(sample_means), max(sample_means), 200)
ax.plot(xr, norm.pdf(xr, np.mean(sample_means), np.std(sample_means)),
        "r-", lw=2, label="Normal (CLT prediction)")
ax.set(title="Central Limit Theorem — Sampling Distribution of Mean Delay",
       xlabel="Sample Mean (min)", ylabel="Density")
ax.legend(); plt.tight_layout(); plt.savefig("prob_02_CLT.png"); plt.show()

# Monte Carlo simulation using lognormal fit on positive-only values
y_positive  = y[y > 0].values
best_params = stats.lognorm.fit(y_positive, floc=0)
mc_delays   = stats.lognorm.rvs(*best_params, size=10_000, random_state=42)
print(f"\n  Monte Carlo (10k) P(delay>15) : {(mc_delays>15).mean():.4f}")
print(f"  Monte Carlo 95th-percentile   : {np.percentile(mc_delays,95):.1f} min")


# =============================================================================
# PROB 3 — HYPOTHESIS TESTING
# =============================================================================
print("\n" + "="*65)
print("PROB 3 — HYPOTHESIS TESTING")
print("="*65)

if COL["weather"]:
    groups = [g[TARGET].values for _,g in df.groupby(COL["weather"]) if len(g)>10]
    if len(groups) == 2:
        t, p = ttest_ind(*groups, equal_var=False)
        d    = (groups[0].mean()-groups[1].mean()) / np.sqrt(
               (groups[0].std()**2+groups[1].std()**2)/2)
        print(f"\n  t-test (weather)  : t={t:.3f}, p={p:.4f}, Cohen's d={d:.3f}")
    elif len(groups) > 2:
        F, p = f_oneway(*groups)
        print(f"\n  ANOVA (weather)   : F={F:.3f}, p={p:.4f}")

if COL["day"]:
    day_g = [g[TARGET].values for _,g in df.groupby(COL["day"]) if len(g)>10]
    if len(day_g) > 2:
        F, p = f_oneway(*day_g)
        print(f"  ANOVA (day)       : F={F:.3f}, p={p:.4f}  "
              f"{'Significant' if p<0.05 else 'Not significant'}")

if COL["weather"]:
    ct2              = pd.crosstab(df[COL["weather"]], df["is_late"])
    chi2, p, dof, _  = chi2_contingency(ct2)
    print(f"  Chi-square        : chi2={chi2:.3f}, p={p:.4f}, dof={dof}")

ci_se  = y.std() / np.sqrt(len(y))
t_crit = stats.t.ppf(0.975, df=len(y)-1)
print(f"  95% CI mean delay : [{y.mean()-t_crit*ci_se:.2f}, "
      f"{y.mean()+t_crit*ci_se:.2f}] min")


# =============================================================================
# FEATURE ENGINEERING
# =============================================================================
print("\n" + "="*65)
print("FEATURE ENGINEERING")
print("="*65)

features = pd.DataFrame(index=df.index)

# base numeric features ordered by correlation strength with the target
numeric_feats = [
    "Distance_KM",
    "Avg_Speed",
    "Fuel_Stops",
    "Traffic_Score",
    "Weather_Impact",
    "Stops",
    "Maintenance_Score",
    "Peak_Hour",
    "Driver_Experience_Years",
    "Road_Quality_Index",
    "Special_Event",
    "Road_Condition",
    "Bus_Age_Years",
    "Passenger_Load",
    "Departure_Hour",
    "Weekend",
]
for col in numeric_feats:
    if col in df.columns:
        features[col] = pd.to_numeric(df[col], errors="coerce").fillna(df[col].median())

# one-hot encode categorical columns
for key in ["weather", "day", "bus_type", "road_type", "traffic"]:
    if COL.get(key) and df[COL[key]].dtype == object:
        dummies  = pd.get_dummies(df[COL[key]], prefix=key, drop_first=True, dtype=float)
        features = pd.concat([features, dummies], axis=1)

features.fillna(0, inplace=True)
features = features.loc[:, features.std() > 0]
print(f"  Base features (before VIF): {features.shape[1]}")

# VIF check on base features before adding any interactions
vif_df = pd.DataFrame({
    "feature": features.columns,
    "VIF"    : [variance_inflation_factor(features.values.astype(float), i)
                for i in range(features.shape[1])]
})
print(f"\n  VIF table (top 10 by VIF):")
print(vif_df.nlargest(10, "VIF").to_string(index=False))

# features with strong correlations are kept regardless of VIF
PROTECTED = {"Distance_KM", "Avg_Speed", "Fuel_Stops",
             "Traffic_Score", "Weather_Impact"}
vif_ok = vif_df[
    (vif_df["VIF"] <= 10) | (vif_df["feature"].isin(PROTECTED))
]["feature"].tolist()
features = features[vif_ok]
print(f"\n  Features after VIF filter ({len(vif_ok)}): {vif_ok}")

# interaction terms are added after VIF so they do not inflate the scores
rain_col  = next((c for c in features.columns if "weather_Rain"  in c), None)
storm_col = next((c for c in features.columns if "weather_Storm" in c), None)

if rain_col and "Peak_Hour" in features.columns:
    features["rain_x_peak"]    = features[rain_col] * features["Peak_Hour"]
if storm_col and "Peak_Hour" in features.columns:
    features["storm_x_peak"]   = features[storm_col] * features["Peak_Hour"]
if "Distance_KM" in features.columns and "Weather_Impact" in features.columns:
    features["dist_x_weather"] = (features["Distance_KM"] * features["Weather_Impact"]) / 1000
stop_col = "Stops" if "Stops" in features.columns else None
traf_col = "Traffic_Score" if "Traffic_Score" in features.columns else None
if stop_col and traf_col:
    features["stops_x_traffic"] = features[stop_col] * features[traf_col]
if "Bus_Age_Years" in features.columns and "Maintenance_Score" in features.columns:
    features["age_x_maintenance"] = (features["Bus_Age_Years"] * (10 - features["Maintenance_Score"])) / 10
if "Fuel_Stops" in features.columns and "Distance_KM" in features.columns:
    features["fuel_x_dist"] = (features["Fuel_Stops"] * features["Distance_KM"]) / 1000

print(f"  Final feature count (with interactions): {features.shape[1]}")
print(f"  Final feature list: {features.columns.tolist()}")

# skewness is near zero so raw target is used — no log transform needed
y_raw = df[TARGET].values.astype(float)
X     = features.values.astype(float)

X_tr, X_te, y_tr, y_te = train_test_split(
    X, y_raw, test_size=0.15, random_state=42)
X_tr, X_val, y_tr, y_val = train_test_split(
    X_tr, y_tr, test_size=0.176, random_state=42)

scaler  = StandardScaler()
X_tr_s  = scaler.fit_transform(X_tr)
X_val_s = scaler.transform(X_val)
X_te_s  = scaler.transform(X_te)
print(f"\n  Train: {len(X_tr):,}  Val: {len(X_val):,}  Test: {len(X_te):,}")
print(f"  Feature matrix shape: {X_tr.shape}")


# =============================================================================
# MODEL TRAINING AND COMPARISON
# =============================================================================
print("\n" + "="*65)
print("MODEL TRAINING AND COMPARISON")
print("="*65)

kf = KFold(n_splits=5, shuffle=True, random_state=42)

def evaluate(name, model, X_test, y_test, fit_scaled=True):
    Xtr = X_tr_s if fit_scaled else X_tr
    model.fit(Xtr, y_tr)
    preds = model.predict(X_test)
    rmse  = np.sqrt(mean_squared_error(y_test, preds))
    mae   = mean_absolute_error(y_test, preds)
    r2    = r2_score(y_test, preds)
    cv    = -cross_val_score(model, Xtr, y_tr, cv=kf,
                              scoring="neg_root_mean_squared_error").mean()
    print(f"  {name:<28} RMSE={rmse:.4f}  MAE={mae:.4f}  "
          f"R2={r2:.4f}  CV-RMSE={cv:.4f}")
    return model, preds, rmse

X_ols     = add_constant(X_tr_s)
ols_model = OLS(y_tr, X_ols).fit()
print("\n  OLS — top coefficients by significance:")
n_params  = len(ols_model.params) - 1
coef_df   = pd.DataFrame({
    "coef":    ols_model.params[1:n_params+1],
    "p-value": ols_model.pvalues[1:n_params+1]
}, index=features.columns[:n_params])
print(coef_df.sort_values("p-value").head(8).to_string())

print()
ridge_cv = RidgeCV(alphas=np.logspace(-3, 3, 50), cv=kf)
lasso_cv = LassoCV(alphas=np.logspace(-3, 1, 50), cv=kf, max_iter=5000)

gb = GradientBoostingRegressor(
    n_estimators=600,
    max_depth=3,
    learning_rate=0.03,
    subsample=0.8,
    min_samples_leaf=15,
    max_features="sqrt",
    random_state=42
)

xgb = XGBRegressor(
    n_estimators=700,
    max_depth=4,
    learning_rate=0.03,
    subsample=0.8,
    colsample_bytree=0.7,
    colsample_bylevel=0.7,
    reg_alpha=0.05,
    reg_lambda=1.0,
    min_child_weight=5,
    gamma=0.1,
    random_state=42,
    verbosity=0
)

_, _,         _ = evaluate("Ridge Regression",   ridge_cv, X_te_s, y_te)
_, _,         _ = evaluate("Lasso Regression",   lasso_cv, X_te_s, y_te)
gb_model,  gb_preds,  _ = evaluate("Gradient Boosting", gb,  X_te, y_te, False)
xgb_model, xgb_preds, _ = evaluate("XGBoost",           xgb, X_te, y_te, False)

best_preds = (gb_preds + xgb_preds) / 2
ens_rmse   = np.sqrt(mean_squared_error(y_te, best_preds))
ens_r2     = r2_score(y_te, best_preds)
print(f"\n  {'Ensemble (GB + XGB)':<28} RMSE={ens_rmse:.4f}  R2={ens_r2:.4f}  <- BEST")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].scatter(y_te, best_preds, alpha=0.3, s=8, color="#5B8DB8")
axes[0].plot([0, y_te.max()], [0, y_te.max()], "r--", lw=1.5)
axes[0].set(title="Predicted vs Actual (Ensemble)",
            xlabel="Actual (min)", ylabel="Predicted (min)")
residuals = y_te - best_preds
axes[1].scatter(best_preds, residuals, alpha=0.3, s=8, color="#E07B54")
axes[1].axhline(0, color="red", lw=1.5, ls="--")
axes[1].set(title="Residuals vs Fitted", xlabel="Fitted", ylabel="Residual")
plt.tight_layout(); plt.savefig("model_residuals.png"); plt.show()

importance = (pd.Series(xgb_model.feature_importances_, index=features.columns)
                .sort_values(ascending=True))
fig, ax = plt.subplots(figsize=(8, max(4, len(importance)*0.35)))
importance.plot.barh(ax=ax, color="#5B8DB8", edgecolor="white")
ax.set(title="Feature Importance — XGBoost", xlabel="Score")
plt.tight_layout(); plt.savefig("model_feature_importance.png"); plt.show()


# =============================================================================
# PREDICTION ENGINE
# =============================================================================
print("\n" + "="*65)
print("PREDICTION ENGINE")
print("="*65)

def predict_delay(user_input: dict,
                  model=xgb_model,
                  feat_cols=features.columns) -> dict:
    """
    Pass a dict of feature values. Keys must match engineered column names.
    Categorical dummies use the encoded format e.g. weather_Rain, day_Fri.
    Numeric features e.g. Distance_KM, Traffic_Score are passed as-is.

    Example:
        predict_delay({"weather_Rain": 1, "Distance_KM": 280,
                       "Traffic_Score": 7, "Peak_Hour": 1})
    """
    row = {c: 0.0 for c in feat_cols}
    for k, v in user_input.items():
        if k in row:
            row[k] = float(v)
    X_pred = np.array([list(row.values())])
    pred   = model.predict(X_pred)[0]

    # bootstrap prediction interval from test-set residuals
    train_r      = y_te - model.predict(X_te)
    boots        = pred + np.random.choice(train_r, 2000, replace=True)
    pi_lo, pi_hi = np.percentile(boots, [2.5, 97.5])

    return {
        "predicted_delay_min"    : round(float(pred), 1),
        "95_PI"                  : (round(max(0, pi_lo), 1), round(pi_hi, 1)),
        "p_more_than_15min_late" : round(float((boots > 15).mean()), 4),
        "p_on_time"              : round(float((boots <= 5).mean()),  4),
    }

# worst-case scenario: storm, peak hour, long route, old poorly-maintained bus
demo_input = {
    "Distance_KM"            : 1400,
    "Avg_Speed"              : 52.0,
    "Fuel_Stops"             : 2,
    "Traffic_Score"          : 6.5,
    "Weather_Impact"         : 6.0,
    "Stops"                  : 10,
    "Maintenance_Score"      : 3.0,
    "Peak_Hour"              : 1,
    "Driver_Experience_Years": 2,
    "Road_Quality_Index"     : 4,
    "Special_Event"          : 1,
    "Road_Condition"         : 3,
    "Bus_Age_Years"          : 14,
    "Passenger_Load"         : 55,
    "Departure_Hour"         : 18,
    "Weekend"                : 0,
    "weather_Storm"          : 1,
    "day_Fri"                : 1,
    "rain_x_peak"            : 0,
    "storm_x_peak"           : 1,
    "dist_x_weather"         : (1400 * 6.0) / 1000,
    "stops_x_traffic"        : 10 * 6.5,
    "age_x_maintenance"      : (14 * (10 - 3.0)) / 10,
    "fuel_x_dist"            : (2 * 1400) / 1000,
}
demo = predict_delay(demo_input)
print(f"\n  Demo — Storm + Friday + Peak Hour + 1400 km + old poorly-maintained bus:")
for k, v in demo.items():
    print(f"    {k:<35} {v}")

print("\n" + "="*65)
print("ALL DONE — Plots saved as PNG files in working directory")
print("="*65)
```
